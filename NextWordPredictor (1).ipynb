```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
df = df['original_title']
```

```python
movie_name = df.to_list()
```

```python
movie_name
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(movie_name)
```

```python
tokenizer.word_index
```

```python
total_unique_words = len(tokenizer.word_index)
total_unique_words
```

```python
seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq
```

```python
X = []
Y = []
for i in seq:
  if len(i)>1:
      for index in range(1,len(i)):
        X.append(i[:index])
        Y.append(i[index])
```

```python
max = 1
for i in range(len(X)):
  a = len(X[i])
  if a>max:
    max = a
print(max)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_X = pad_sequences(X, maxlen = max, padding='pre')
```

```python
padded_X[0]
```

```python
from tensorflow.keras.utils import to_categorical
Y = to_categorical(Y,num_classes=total_unique_words+1)
```

```python
print(padded_X.shape)
print(Y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(5045, 100, input_length=14))
model.add(LSTM(150))
model.add(Dense(5045, activation='softmax'))
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(padded_X,Y,epochs=100,batch_size=32)
```

```python
import time
text = "Pirates"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=14, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
